# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv('../data/liquor_store_sales.csv')
data.head()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
df_Items = data[['ItemType', 'RetailSales', 'RetailTransfers', 'WarehouseSales']]
df_Items.set_index('ItemType')
df_Items_grouped = df_Items.groupby(['ItemType']).sum()
figure = df_Items_grouped.plot.bar(rot=0)
figure.xaxis.set_tick_params(rotation=45);
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
data['TotalSales'] = data['WarehouseSales']+data['RetailSales']
df_suppliers = data[['Supplier','TotalSales']]
df_suppliers_grouped = df_suppliers.groupby(['Supplier']).sum()
df_suppliers_grouped_sort = df_suppliers_grouped.sort_values(by=['TotalSales'], ascending = False)
Top_Suppliers = df_suppliers_grouped_sort.head(10)
Top_Suppliers.reset_index()
Top_Suppliers.plot.barh(rot=0);
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
df = data[['Year', 'Month', 'RetailSales', 'RetailTransfers', 'WarehouseSales']]
df = df.astype({"Year": str, "Month": str})


df['YearMonth'] = df['Month'] + '_' + df['Year']
df_YMonth = df[['YearMonth', 'RetailSales', 'RetailTransfers', 'WarehouseSales']]
df_YMonth_grouped = df_YMonth.groupby(['YearMonth']).sum()
df_YMonth_grouped.reset_index()
count = [6,7,8,9,10,1,2,3,4,5]
df_YMonth_grouped['count'] = count
df_YMonth_grouped_sort = df_YMonth_grouped.sort_values(by =['count'])
df_YMonth_grouped_sort = df_YMonth_grouped_sort [['RetailSales', 'RetailTransfers', 'WarehouseSales']]
df_YMonth_grouped_sort.plot(rot=0);
```

## 4. Plot the same information as above but as a bar chart.

```python
figure = df_YMonth_grouped_sort.plot.bar(rot=0)
figure.xaxis.set_tick_params(rotation=45)
figure.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), ncol=1);
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
df4 = data[['Year', 'Month', 'RetailSales', 'ItemType']]
df4 = df4.astype({"Year": str, "Month": str})

df4['YearMonth'] = df4['Month'] + '_' + df4['Year']
df4 = df4[['YearMonth', 'ItemType','RetailSales']]
df5 = df4.groupby(['YearMonth','ItemType']).sum()
df6 = df5.reset_index()
df7 = df6.replace(['4_2017', '5_2017', '6_2017', '7_2017', '8_2017', '9_2017', '10_2017', '11_2017', '12_2017','1_2018', '2_2018'], [2017.04, 2017.05, 2017.06, 2017.07, 2017.08, 2017.09, 2017.10, 2017.11,2017.12, 2018.01, 2018.02])
df8 = df7.sort_values(by=['YearMonth'])


fig = plt.figure(figsize=(10,6))

ax = sns.pointplot(x='YearMonth', y='RetailSales', data=df8, hue='ItemType', size = 'ItemType', err_style = 'bars')

ax.legend(loc = 'center left', bbox_to_anchor= (1.0,0.5), ncol=1);
```

## 6. Plot the same information as above but as a bar chart.

```python
fig = plt.figure(figsize=(12, 8))
ax = sns.barplot(x='YearMonth', y='RetailSales', hue='ItemType', data=df8)
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), ncol=1);
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
figure = sns.lmplot(x="RetailSales", y="RetailTransfers", data=data);
```
